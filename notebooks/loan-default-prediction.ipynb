```python
def drop_cols_by_missing(df, threshold=0.5, return_dropped=False, verbose=True):
    """
    Efficiently drop columns exceeding missing value threshold.
    
    Parameters
    ----------
    df : pandas.DataFrame
    threshold : float
        Fraction of missing values (0–1)
    return_dropped : bool
        If True, also returns list of dropped columns
    verbose : bool
        Print summary
    
    Returns
    -------
    DataFrame (and optionally list of dropped columns)
    """
    
    # Compute missing fraction ONCE (fast vectorized op)
    missing_frac = df.isna().mean()

    # Boolean mask (faster than building lists first)
    drop_mask = missing_frac > threshold

    # Select columns to drop
    cols_to_drop = missing_frac.index[drop_mask]
    print(cols_to_drop)

    if verbose:
        print(f"Dropping {len(cols_to_drop)} columns > {threshold:.0%} missing")

    # Drop in one operation (fastest path in pandas)
    df_clean = df.drop(columns=cols_to_drop)

    if return_dropped:
        return df_clean, cols_to_drop.tolist()

    return df_clean
```

```python
def drop_rows_based_on_column_missing(df, col_missing_threshold=0.05, verbose=True):
    """
    Drop rows based on missingness in feature columns only.
    NEVER uses 'dti' in any cleaning logic.
    """

    # Step 1: compute column missing rates
    col_missing = df.isna().mean()

    # Step 2: select "good columns" BUT EXCLUDE target explicitly
    excluded_cols = ["dti"]
    good_cols = [
        col for col in col_missing.index
        if col_missing[col] <= col_missing_threshold
        and col not in excluded_cols
    ]

    # Step 3: drop rows with NA in ONLY feature columns
    cleaned_df = df.dropna(subset=good_cols)

    if verbose:
        print(f"Using {len(good_cols)} feature columns")
        print(f"Dropped {len(df) - len(cleaned_df)} rows (features only, not dti-aware)")

    return cleaned_df

df = drop_rows_based_on_column_missing(df, col_missing_threshold=0.03)
```

```python
df["revol_util_missing"] = df["revol_util"].isna().astype(int)

df["revol_util"] = df["revol_util"].fillna(df["revol_util"].median())
```

```python
df["all_util_missing"] = df["all_util"].isna().astype(int)

df["all_util"] = df["all_util"].fillna(df["all_util"].median())
```

```python
# TYPE 1: counts
cols_type1 = [
    "open_il_24m", "open_il_12m", "open_rv_12m",
    "open_acc_6m", "inq_fi", "inq_last_12m", "total_cu_tl"
]

for col in cols_type1:
    df[col + "_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(0)

# TYPE 2: balances
cols_type2 = ["total_bal_il", "max_bal_bc"]

for col in cols_type2:
    df[col + "_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(df[col].median())
```

```python
# -----------------------------
# Time features
# -----------------------------

time_cols = [
    "mo_sin_old_rev_tl_op",
    "mo_sin_rcnt_rev_tl_op",
    "mo_sin_rcnt_tl"
]

for col in time_cols:
    df[f"{col}_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(999)


# -----------------------------
# Count features
# -----------------------------

count_cols = [
    "num_rev_accts",
    "num_op_rev_tl",
    "num_il_tl",
    "num_rev_tl_bal_gt_0",
    "num_tl_op_past_12m",
    "num_tl_90g_dpd_24m",
    "num_tl_30dpd",
    "num_accts_ever_120_pd",
    "num_actv_rev_tl",
    "num_bc_tl",
    "num_actv_bc_tl"
]

for col in count_cols:
    df[f"{col}_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(0)


# -----------------------------
# Money / balance features
# -----------------------------

money_cols = [
    "tot_cur_bal",
    "total_rev_hi_lim",
    "total_il_high_credit_limit",
    "avg_cur_bal",
    "tot_hi_cred_lim",
    "tot_coll_amt"
]

for col in money_cols:
    df[f"{col}_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(df[col].median())


# -----------------------------
# Percentage features
# -----------------------------

pct_cols = [
    "pct_tl_nvr_dlq"
]

for col in pct_cols:
    df[f"{col}_missing"] = df[col].isna().astype(int)
    df[col] = df[col].fillna(df[col].median())


# Verify
print(df[time_cols + count_cols + money_cols + pct_cols].isna().sum().sum())
```
